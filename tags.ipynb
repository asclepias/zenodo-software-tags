```python
import json
from collections import Counter
import re
```

```python
# Load the raw tags dump
with open('tags.json', 'r') as fp:
    tags = json.load(fp)
```

```python
# Regular expression which will match the tags
re_semver_pat = \
('^((v|V|(v\.)|(V\.))?)(0|[1-9]\d*)((\.(0|[1-9]\d*))?)((\.(0|[1-9]\d*))?)((\.(0|[1-9]\d*))?)'
 '(-(0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*)(\.(0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*))*)?(\+[0-9a-zA-Z-]+(\.[0-9a-zA-Z-]+)*)?$')
re_sv =re.compile(re_semver_pat)

# Some examples of tags which are matched by the expression above
good_test = ['1', '0', 'v1', 'V1', 'V1.0', '1.0', '0.5', '1.0.0',
             '0.1.0', 'v1.0.0', '2017-01-01', '1.0.0.0', 'v.0.1',
             '1.0.0-alpha1', '1-rc1', '1.0-dev2017.1.10']
# Examples of tags which are not
bad_test = ['x', '0.x.0', 'x.0', 'x', 'foobar', 'x-alpha2', '1.0.0a2', '1.01']
```

```python
# This cell should print nothing
for t in good_test:
    if not re_sv.match(t):
        print(t, "GOOD")
for t in bad_test:
    if re_sv.match(t):
        print(t, "BAD")
```

```python
tags_match = [(t, bool(re_sv.match(t))) for t in tags]
    
```

```python
cnts = Counter(t[1] for t in tags_match)  # Count the number of tags which are semver-like (True) vs not (False) 
print(cnts)
print(float(cnts[True]/len(tags_match)))  # Fraction that matches
```

```python
# Print the tags which do not match
for t in tags_match:
    if not t[1]:
        print(t[0])
```
